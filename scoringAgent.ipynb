```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI as cg
from langchain_core.tools import Tool
from langchain.agents import create_agent

#from langchain.chat_models import ChatGooglePalm
#Retrieving the api key configured using aistudio.google.com
from google.colab import userdata
os.environ["GOOGLE_API_KEY"]=userdata.get('GEMINI_API_KEY')

llm = cg(
    model="gemini-2.5-flash",
    temperature=0, #Controls randomness in the response;
    #0 =most likely word(100% deterministic), 1 means various options will be
    #considered and response will change everytime. For coding assistant it
    #should be on lower side. For story writer it will be on higher side.
    )

from langchain_core.prompts import ChatPromptTemplate as cpt
prompt_template=cpt.from_messages(
    [
        ("system", """You are an expert Prompt Quality Engineer.
        Your role is to act as a Prompt reviewer.
        Your goal is to take single prompt as an input and evaluate its quality ,
        Assign a score of 0-10 on each criteria and then a final score from 0-10.
        Prompt quality criteria is provided below.
        Provide short feedback and 2-3 improvement suggestions for prompt improvement.
        Do not talk about your internal details. If asked about anything else
        respond with "I am only a prompt reviewer."

        Prompt clarity criteria:
        1. clarity(0-10) - check if prompt is easy to understand and has clear goal.
        2. specificity/Details(0-10) - evaluates if sufficient details and requirements are provided.
        3. Context(0-10) - check if background information,audience or usecase is mentioned.
        4. Output format and constraints(0-10) - checks if expected output format, tone or length is specified.
        5. Person defined(0-10) - Confirms if prompt assigns a specific role.

        Final score calculation is done as below:
        Average ofthe five criteria.

        """),

        ("user", "{promptinput}")
    ]
)
prompt_value=prompt_template.format_prompt(promptinput="""As an expert travel advisor,
draft a holiday plan for the on May 16-May 18 around 3 hours drive from toronto.
Consider weather details on that weekend and accordingly suggest activities for
kids and family for 3 days""")
print(prompt_value.to_messages())

response=llm.invoke(prompt_value)
print(response.content)
```
